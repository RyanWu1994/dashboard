```python
import pandas as pd
import numpy as np
import pymysql

from sqlalchemy import create_engine
from collections import Counter
```

```python
def mysql_to_df():
    db = pymysql.connect("192.168.22.110","root","Pn123456","17CE" )
    # prepare a cursor object using cursor() method
    cursor = db.cursor()

    # Prepare SQL query to INSERT a record into the database.
    sql = "SELECT * from main_information WHERE DATE > date_sub(now(), INTERVAL 64 hour)"
    try:
       # Execute the SQL command
        cursor.execute(sql)
        result = cursor.fetchall()
        Date = []
        Line = []
        Fastest_node = []
        Fastest_node_seconds = []
        Slowest_node = []
        Slowest_node_seconds = []
        Average_response = []

        for i in range(len(result)):
            for index,content in enumerate(result[i]):
                if index == 0:
                    URL = content
                elif index == 1:
                    Date.append(content.strftime('%Y-%m-%d %H:%M:%S'))
                elif index == 2:
                    Line.append(content)
                elif index == 3:
                    Fastest_node.append(content)
                elif index == 4:
                    Fastest_node_seconds.append(content)
                elif index == 5:
                    Slowest_node.append(content)
                elif index == 6:
                    Slowest_node_seconds.append(content)
                elif index == 7:
                    Average_response.append(content)
                    
        dict_result = {
            "URL":URL,
            "Date":Date,
            "Line":Line,
            "Fastest_node":Fastest_node,
            "Fastest_node_seconds":Fastest_node_seconds,
            "Slowest_node":Slowest_node,
            "Slowest_node_seconds":Slowest_node_seconds,
            "Average_response":Average_response
        }

        db.commit()
        
    except:

        db.rollback()
        
    db.close()
    
    return dict_result

result = mysql_to_df()
```
